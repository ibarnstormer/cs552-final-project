```python
import numpy as np
import matplotlib.pyplot as plt

# Load the loss histories
conv_vae_history = np.load('convolutional_vae_loss_history.npy', allow_pickle=True).item()
vanilla_vae_history = np.load('vanilla_vae_loss_history.npy', allow_pickle=True).item()

# Choose which loss components to plot
loss_components = ['total_loss', 'reconstruction_loss', 'kl_divergence']

# Plot
plt.figure(figsize=(12, 8))
for i, component in enumerate(loss_components, 1):
    plt.subplot(len(loss_components), 1, i)
    plt.plot(conv_vae_history[component], label='Convolutional VAE', color='blue')
    plt.plot(vanilla_vae_history[component], label='Vanilla VAE', color='orange')
    plt.title(component.replace('_', ' ').title())
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load all four loss histories
vq_vae = np.load('vq-vae_loss_history.npy', allow_pickle=True).item()
vq_vtae = np.load('vq-vtae_loss_history.npy', allow_pickle=True).item()
vq_vae_2 = np.load('vq-vae-2_loss_history.npy', allow_pickle=True).item()
vq_vtae_2 = np.load('vq-vtae-2_loss_history.npy', allow_pickle=True).item()

# Dictionary of models for easier looping
models = {
    'VQ-VAE': vq_vae,
    'VQ-VTAE': vq_vtae,
    'VQ-VAE-2': vq_vae_2,
    'VQ-VTAE-2': vq_vtae_2
}

# Set of loss components to compare
loss_components = ['total_loss', 'reconstruction_loss', 'perplexity', 'latent_loss']

# Plot
plt.figure(figsize=(14, 18))
for i, component in enumerate(loss_components, 1):
    plt.subplot(len(loss_components), 1, i)
    
    for model_name, history in models.items():
        if component in history:
            plt.plot(history[component], label=model_name)
    
    plt.title(component.replace('_', ' ').title())
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
